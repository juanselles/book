# Classification - Decision Tree example

Simple classifier using sklearn's DecisionTreeClassifier.

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create synthetic dataset
np.random.seed(0)
X = np.vstack([np.random.normal(loc=0, scale=1, size=(50,2)),
               np.random.normal(loc=3, scale=1, size=(50,2))])
Y = np.array([0]*50 + [1]*50)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)
print(classification_report(y_test, clf.predict(X_test)))

plt.figure(figsize=(8,6))
plot_tree(clf, feature_names=['f1','f2'], class_names=['class0','class1'], filled=True)
plt.show()
```
