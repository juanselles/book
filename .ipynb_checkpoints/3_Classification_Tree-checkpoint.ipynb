# Classification - Decision Tree example

Simple classifier using sklearn's DecisionTreeClassifier.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 1. Crear datos de entrenamiento sintéticos y fáciles de entender
# (Altura en cm, Peso en kg, Edad en años)
datos = {
    'Altura': [25, 30, 45, 50, 60, 22, 28],
    'Peso': [4, 5, 15, 20, 30, 3, 6],
    'Edad': [2, 3, 5, 4, 6, 1, 2],
    'Animal': ['Gato', 'Gato', 'Perro', 'Perro', 'Perro', 'Gato', 'Gato']
}
df = pd.DataFrame(datos)

# Separar características (X) de la etiqueta a predecir (y)
X = df[['Altura', 'Peso', 'Edad']]
y = df['Animal']

# 2. Crear y entrenar el modelo (el "Cerebro")
arbol = DecisionTreeClassifier(max_depth=3, random_state=42)
arbol.fit(X, y)

# 3. Predecir un nuevo animal misterioso 
# (Ejemplo: 40cm de altura, 12kg de peso, 4 años)
prediccion = arbol.predict([[40, 12, 4]])
print(f"Predicción para el nuevo animal [40cm, 12kg, 4 años]: ¡Es un {prediccion[0]}!")

# 4. ¡DIBUJAR EL ÁRBOL! (La parte divertida)
plt.figure(figsize=(10,6))
plot_tree(arbol, 
          filled=True, 
          feature_names=['Altura', 'Peso', 'Edad'], 
          class_names=['Gato', 'Perro'],
          rounded=True)
plt.show()
```
