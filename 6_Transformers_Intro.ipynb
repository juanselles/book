# Transformers - Intro example

Simple example showing tokenization and using a small transformer model with Hugging Face (if installed).

```python
try:
    from transformers import pipeline
    summarizer = pipeline('summarization')
    print('Hugging Face transformers available. Example:')
    print(summarizer('Transformers are models that use attention to process sequences.', max_length=30, min_length=5))
except Exception as e:
    print('Transformers not available in the environment. This cell shows the code you can run locally or on Colab where transformers is installed.')
    print(e)

# Tokenization example with Hugging Face (commented out if not installed)
# from transformers import AutoTokenizer
# tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
# tokens = tokenizer('This is a short example of tokenization', return_tensors='pt')
# print(tokens)
```
